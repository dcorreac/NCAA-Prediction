```python
import pandas
from sklearn.model_selection import cross_val_score
import numpy
from sklearn.dummy import DummyClassifier

basketball = pandas.read_csv('cleanNCAA.csv')
```

```python

predictorList = ['Assists Per Game', 'Turnovers Per Game', 'Personal Fouls Per Game', 'Steals Per Game', 'Free-Throw Percentage', 'Field-Goal Percentage', 'Three-Point Field-Goal Percentage', 'Turnover Margin', 'Blocked Shots Per Game', 'Scoring Margin', 'Won-Lost Percentage', 'Three Pt FG Defense', 'Scoring Offense', 'Scoring Defense', 'Three-Point Field Goals Per Game', 'Rebound Margin', 'Field-Goal Percentage Defense']
predictor_data = basketball[predictorList]
predictor_data
```

```python
outcome_data = basketball['Qualified'] #subset the dataset to isolate just the outcome and call it outcome_data
outcome_data = outcome_data.values.ravel() #you have flatten out the outcome for sklearn to parse it properly
outcome_data #view the outcome data
```

```python
from sklearn.ensemble import RandomForestClassifier




n_estimators = 30

max_features = 8

max_depth = 8

min_samples_split = 10

bootstrap = False




model = RandomForestClassifier(n_estimators=n_estimators, 
                              max_features= max_features, 
                              max_depth = max_depth, 
                              min_samples_split = min_samples_split,
                              bootstrap=bootstrap)
model
```

```python

folds = 5
evaluation_metric = "accuracy"

scores = cross_val_score(model, predictor_data, outcome_data, cv=folds,scoring=evaluation_metric)
accuracy = scores.mean()
```

```python
from sklearn.metrics import recall_score

predictions = model.fit(predictor_data, outcome_data)
predictionsResult = predictions.predict(predictor_data)

recall = recall_score(outcome_data, predictionsResult, average = 'binary', pos_label = 'Yes')
```

```python
from sklearn.metrics import precision_score
precision = precision_score(outcome_data, predictionsResult, average = 'binary', pos_label = 'Yes')
```

```python
print('Random-----------Forest-----------Classifier')
print('Recall:')
print(recall)
print('Precision:')
print(precision)
print('Accuracy:')
print(accuracy)
```

```python
from sklearn.model_selection import cross_val_score
import numpy
from sklearn.dummy import DummyClassifier

scores = cross_val_score(DummyClassifier(), predictor_data, outcome_data, cv=folds,scoring=evaluation_metric)
scores.mean()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import f1_score

clf = AdaBoostClassifier(n_estimators = 3000)
adaPredictions = clf.fit(predictor_data, outcome_data)
adaPredictions = adaPredictions.predict(predictor_data)

recall = recall_score(outcome_data, adaPredictions, average = 'binary', pos_label = 'Yes')
precision = precision_score(outcome_data, adaPredictions, average = 'binary', pos_label = 'Yes')

folds = 10
evaluation_metric = "accuracy"

scores = cross_val_score(clf, predictor_data, outcome_data, cv=folds,scoring=evaluation_metric)
accuracy = scores.mean()

print('--------------ADA----------BOOST-----')
print('Recall:')
print(recall)
print('Precision:')
print(precision)
print('Accuracy:')
print(accuracy)
print('F1 score:')
print(f1_score(outcome_data, adaPredictions, average = 'weighted'))
```

```python
basketball['adaPredictions'] = adaPredictions
```
